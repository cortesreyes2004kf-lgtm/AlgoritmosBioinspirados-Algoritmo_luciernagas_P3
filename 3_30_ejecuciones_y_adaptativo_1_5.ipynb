# 1. Algoritmo de las Luciérnagas con adaptación dinámica (Regla 1/5)

El siguiente código se desarrolla con el fin de cumplir la consigna de **adaptación de parámetros en línea según el número de éxitos**.

Ahora en lugar de reducir la aleatoriedad (α) por fórmula matemática, se utiliza la *Regla de 1/5 de éxito*.

* Si en la generación actual más del 20% de las luciérnagas mejoran su fitness (éxito), se aumenta el tamaño de paso α para explorar más rápido.

* Si menos del 20% mejoraron, se disminuye α para explotar la zona cercana.

```python
# Importamos NumPy para operaciones matemáticas vectorizadas y eficientes
import numpy as np
# Importamos Matplotlib para generar visualizaciones y gráficas de convergencia
import matplotlib.pyplot as plt

# ==========================================================
# 1. FUNCIONES BENCHMARK
# ==========================================================

# Función matemática de Ackley
def ackley(x):
    d = len(x) # Dimensión del problema
    sum_sq = np.sum(x**2) # Sumatoria de cuadrados
    sum_cos = np.sum(np.cos(2 * np.pi * x)) # Sumatoria de cosenos
    # Ecuación completa de Ackley
    return -20 * np.exp(-0.2 * np.sqrt(sum_sq / d)) - np.exp(sum_cos / d) + 20 + np.exp(1)

# Función matemática de Griewank
def griewank(x):
    d = len(x) # Dimensión del problema
    sum_term = np.sum(x**2) / 4000 # Sumatoria penalizada
    indices = np.arange(1, d + 1) # Arreglo de índices [1, 2, ..., d]
    prod_term = np.prod(np.cos(x / np.sqrt(indices))) # Productoria de cosenos escalados
    # Ecuación completa de Griewank
    return sum_term - prod_term + 1

# Función matemática de Rastrigin
def rastrigin(x):
    d = len(x) # Dimensión
    # Ecuación de Rastrigin (altamente multimodal)
    return 10 * d + np.sum(x**2 - 10 * np.cos(2 * np.pi * x))

# Función matemática de Rosenbrock
def rosenbrock(x):
    # Ecuación del valle de Rosenbrock iterando sobre pares de dimensiones
    return np.sum(100.0 * (x[1:] - x[:-1]**2)**2.0 + (x[:-1] - 1)**2.0)

# ==========================================================
# 2. ALGORITMO DE LUCIÉRNAGAS ADAPTATIVO (Regla 1/5)
# ==========================================================

# Función para acotar el espacio de búsqueda
def apply_limits(ns, lb, ub):
    ns = np.maximum(ns, lb) # Trunca por debajo del límite inferior
    ns = np.minimum(ns, ub) # Trunca por encima del límite superior
    return ns

# Función principal del FA con adaptación en línea
def adaptive_firefly(func, d, bounds, N=100, G_MAX=5000):
    lb, ub = bounds[0], bounds[1] # Límites del problema
    scale = np.abs(ub - lb) # Escala espacial máxima

    # Parámetros Base
    alpha = 0.5       # Parámetro de salto aleatorio inicial (Se adaptará en línea)
    betamin = 0.2     # Atractividad mínima
    gamma = 1.0       # Coeficiente de absorción de luz

    # Inicialización aleatoria de la población de luciérnagas
    ns = lb + (ub - lb) * np.random.rand(N, d)
    # Vector inicial de "luz" (fitness) seteado a infinito
    lightn = np.ones(N) * np.inf

    history = [] # Lista para almacenar la curva de convergencia

    # BUCLE PRINCIPAL DE GENERACIONES
    for g in range(G_MAX):

        # Evaluamos el fitness de las posiciones actuales
        current_fitness = np.array([func(ind) for ind in ns])

        # --- ADAPTACIÓN DE PARÁMETROS EN LÍNEA (Basado en regla de 1/5 de éxito) ---
        if g > 0:
            # Contamos cuántas luciérnagas mejoraron su fitness respecto a la generación anterior
            exitos = np.sum(current_fitness < lightn)
            tasa_exito = exitos / N # Calculamos el porcentaje de éxito

            # Si hay muchos éxitos (>20%), el algoritmo explora bien. AUMENTAMOS el salto alpha.
            if tasa_exito > 0.2:
                alpha = min(alpha * 1.05, 1.0) # Límite superior de 1.0
            # Si hay pocos éxitos (<20%), el algoritmo no halla mejoras. REDUCIMOS alpha para explotar.
            elif tasa_exito < 0.2:
                alpha = max(alpha * 0.95, 0.001) # Límite inferior de 0.001
            # Si es exactamente 20%, alpha se mantiene igual.

        lightn = current_fitness.copy() # Actualizamos el registro de luz

        # Ordenamos las luciérnagas de la más brillante (menor costo) a la más opaca
        idx = np.argsort(lightn)
        lightn = lightn[idx]
        ns = ns[idx]

        # Guardamos la mejor de esta generación
        history.append(lightn[0])

        # Copias de seguridad para la fase de movimiento
        nso = ns.copy()
        lighto = lightn.copy()

        # MOVIMIENTO HACIA LUCIÉRNAGAS MÁS BRILLANTES
        for i in range(N):
            for j in range(N):
                # Si la luciérnaga j es mejor que i, i se mueve hacia j
                if lighto[j] < lightn[i]:
                    r = np.linalg.norm(ns[i] - nso[j]) # Distancia Euclidiana
                    r_scaled = r / scale # Distancia escalada
                    # Cálculo de atractividad
                    beta = (1.0 - betamin) * np.exp(-gamma * r_scaled**2) + betamin
                    # Ruido aleatorio usando el ALPHA ADAPTATIVO
                    tmpf = alpha * (np.random.rand(d) - 0.5) * scale
                    # Actualización de posición matemática
                    ns[i] = ns[i] * (1 - beta) + nso[j] * beta + tmpf

        # Acotamos a los límites tras los saltos
        ns = apply_limits(ns, lb, ub)

    return ns[0], lightn[0], history

# ==========================================================
# 3. EJECUCIÓN DEL EXPERIMENTO (30 EJECUCIONES)
# ==========================================================
if __name__ == "__main__":
    problemas = {
        "Ackley": (ackley, [-32.768, 32.768]),
        "Griewank": (griewank, [-600, 600]),
        "Rastrigin": (rastrigin, [-5.12, 5.12]),
        "Rosenbrock": (rosenbrock, [-2.048, 2.048])
    }

    D = 10 # 10 variables de decisión
    N_RUNS = 30 #30 ejecuciones independientes
    np.random.seed(42) # Semilla matemática

    for nombre_prob, (funcion, limites) in problemas.items():
        print(f"\n{'='*50}\n Resolviendo: {nombre_prob.upper()}\n{'='*50}")
        resultados_finales = []
        curvas_convergencia = []

        for corrida in range(N_RUNS):
            print(f"-> Ejecución {corrida + 1} de {N_RUNS}...")
            # Llamamos al algoritmo modificado
            mejor_pos, mejor_fit, historial = adaptive_firefly(funcion, D, limites, N=100, G_MAX=5000)
            resultados_finales.append(mejor_fit)
            curvas_convergencia.append(historial)

        resultados_finales = np.array(resultados_finales)
        print(f"\n--- ESTADÍSTICAS ({nombre_prob}) ---")
        print(f"Promedio (Media): {np.mean(resultados_finales):.6e}")
        print(f"Desviación estándar:   {np.std(resultados_finales):.6e}")

        print("\nDatos para prueba de Wilcoxon (30 ejecuciones):")
        print(list(resultados_finales)) # Imprime lista final

        # Graficamos la convergencia
        mediana_curva = np.median(np.array(curvas_convergencia), axis=0)
        plt.figure(figsize=(10, 6))
        plt.plot(mediana_curva, label="FA Adaptativo (Mediana)", color='darkviolet')
        plt.yscale('log')
        plt.title(f"Convergencia {nombre_prob} (30 ejecuciones)")
        plt.xlabel("Generaciones")
        plt.ylabel("Mejor Fitness (Log)")
        plt.grid(True, ls="--", alpha=0.5)
        plt.legend()
        plt.savefig(f"convergencia_{nombre_prob}.png")
        plt.show()
```

# Debido a que el tiempo de ejecución se extendía a un aproximado de 12 horas continuas, se opta mejor por separar cada función para ejecutar una a una y no perder el progreso

```python
# Importamos NumPy para operaciones matemáticas vectorizadas y eficientes
import numpy as np
# Importamos Matplotlib para generar visualizaciones y gráficas de convergencia
import matplotlib.pyplot as plt

# ==========================================================
# 1. FUNCIONES BENCHMARK
# ==========================================================

def ackley(x):
    d = len(x)
    sum_sq = np.sum(x**2)
    sum_cos = np.sum(np.cos(2 * np.pi * x))
    return -20 * np.exp(-0.2 * np.sqrt(sum_sq / d)) - np.exp(sum_cos / d) + 20 + np.exp(1)

def griewank(x):
    d = len(x)
    sum_term = np.sum(x**2) / 4000
    indices = np.arange(1, d + 1)
    prod_term = np.prod(np.cos(x / np.sqrt(indices)))
    return sum_term - prod_term + 1

def rastrigin(x):
    d = len(x)
    return 10 * d + np.sum(x**2 - 10 * np.cos(2 * np.pi * x))

def rosenbrock(x):
    return np.sum(100.0 * (x[1:] - x[:-1]**2)**2.0 + (x[:-1] - 1)**2.0)

# ==========================================================
# 2. ALGORITMO DE LUCIÉRNAGAS ADAPTATIVO (Regla 1/5)
# ==========================================================

def apply_limits(ns, lb, ub):
    ns = np.maximum(ns, lb)
    ns = np.minimum(ns, ub)
    return ns

def adaptive_firefly(func, d, bounds, N=100, G_MAX=5000):
    lb, ub = bounds[0], bounds[1]
    scale = np.abs(ub - lb)

    alpha = 0.5
    betamin = 0.2
    gamma = 1.0

    ns = lb + (ub - lb) * np.random.rand(N, d)
    lightn = np.ones(N) * np.inf
    history = []

    for g in range(G_MAX):
        current_fitness = np.array([func(ind) for ind in ns])

        if g > 0:
            exitos = np.sum(current_fitness < lightn)
            tasa_exito = exitos / N
            if tasa_exito > 0.2:
                alpha = min(alpha * 1.05, 1.0)
            elif tasa_exito < 0.2:
                alpha = max(alpha * 0.95, 0.001)

        lightn = current_fitness.copy()

        idx = np.argsort(lightn)
        lightn = lightn[idx]
        ns = ns[idx]

        history.append(lightn[0])

        nso = ns.copy()
        lighto = lightn.copy()

        for i in range(N):
            for j in range(N):
                if lighto[j] < lightn[i]:
                    r = np.linalg.norm(ns[i] - nso[j])
                    r_scaled = r / scale
                    beta = (1.0 - betamin) * np.exp(-gamma * r_scaled**2) + betamin
                    tmpf = alpha * (np.random.rand(d) - 0.5) * scale
                    ns[i] = ns[i] * (1 - beta) + nso[j] * beta + tmpf

        ns = apply_limits(ns, lb, ub)

    return ns[0], lightn[0], history

# ==========================================================
# 3. FUNCIÓN EJECUTORA
# ==========================================================
def ejecutar_experimento(nombre_prob, funcion, limites, D=10, N_RUNS=30):
    np.random.seed(42) # Semilla independiente para cada celda
    print(f"\n{'='*50}\n Resolviendo: {nombre_prob.upper()}\n{'='*50}")

    resultados_finales = []
    curvas_convergencia = []

    for corrida in range(N_RUNS):
        print(f"-> Ejecución {corrida + 1} de {N_RUNS}...")
        mejor_pos, mejor_fit, historial = adaptive_firefly(funcion, D, limites, N=100, G_MAX=5000)
        resultados_finales.append(mejor_fit)
        curvas_convergencia.append(historial)

    resultados_finales = np.array(resultados_finales)
    print(f"\n--- ESTADÍSTICAS ({nombre_prob}) ---")
    print(f"Promedio (Media): {np.mean(resultados_finales):.6e}")
    print(f"Desviación estándar:   {np.std(resultados_finales):.6e}")

    print("\nDatos para prueba de comparación (30 ejecuciones):")
    print(list(resultados_finales))

    # Gráfica
    mediana_curva = np.median(np.array(curvas_convergencia), axis=0)
    plt.figure(figsize=(10, 6))
    plt.plot(mediana_curva, label="FA Adaptativo (Mediana)", color='darkviolet')
    plt.yscale('log')
    plt.title(f"Convergencia {nombre_prob} (30 ejecuciones)")
    plt.xlabel("Generaciones")
    plt.ylabel("Mejor Fitness (Log)")
    plt.grid(True, ls="--", alpha=0.5)
    plt.legend()
    plt.savefig(f"convergencia_{nombre_prob}.png")
    plt.show()
```
