Matemáticamente ocurre así porque la ED perturba el espacio de búsqueda usando las diferencias geométricas entre los vectores de los individuos lo que es superior cuando se trabajan los espacios continuos y valles estrechos como en la función Rosenbrock.

Por otro lado, el algoritmo de las luciérnagas depende de un mecanismo de atracción que se centraliza hacia el líder; en topologías que son multimodales como Rastrigin que está llena de muchísimos óptimos locales, la atracción lumínica forzó al enjambre a converger de manera prematura en un óptimo local cercano (36.47) todo a pesar de que el mecanismo de auto-adaptación redujo el impacto del estancamiento que sufrieron las EE.
